    if isinstance(Dataset,pd.DataFrame):
        continous=[]
        categorial=[]
        for col in Dataset.columns:

                continous.append(col)
            else:
                categorial.append(col)

##### Missing values checking 

for first let check how to get the percentage of missing values for each attributes

we will use a numpy build in method count non zero and create a new dataset of each attribute and missing vallues percentage

```python
def PercentageMissin(Dataset):
    """this function will return the percentage of missing values in a dataset """
    if isinstance(Dataset,pd.DataFrame):
        adict={} #a dictionary conatin keys columns names and values percentage of missin value in the columns
        for col in Dataset.columns:
            adict[col]=(np.count_nonzero(Dataset[col].isnull())*100)/len(Dataset[col])
        return pd.DataFrame(adict,index=['% of missing'],columns=adict.keys())
    else:
        raise TypeError("can only be used with panda dataframe")
            
```

###### filling missing values

let look to approach for filling mising values given by in anlythics vidha tutorial

according to the documenttion and information about data we will fill the missing value with a radom number between 40-55 and 
will be just for making good visualisation but in future we will explore more methods for filling missing values according to origin

```python
#DatasetSMFCont['PERC1-2ndYear']
DatasetSMFCont['PERC1-2ndYear']=DatasetSMFCont['PERC1-2ndYear'].apply(lambda x: np.random.choice([x for x in range(40,55)]) if (np.isnan(x)) else x)
DatasetSMFCont['PERC1-1year']=DatasetSMFCont['PERC1-1year'].apply(lambda x: np.random.choice([x for x in range(40,55)]) if (np.isnan(x)) else x)
DatasetSMFCont['PERC2-1year']=DatasetSMFCont['PERC2-1year'].apply(lambda x: np.random.choice([x for x in range(40,55)]) if (np.isnan(x)) else x)
DatasetSMFCont['PERC2-2ndYear']=DatasetSMFCont['PERC2-2ndYear'].apply(lambda x: np.random.choice([x for x in range(40,55)]) if (np.isnan(x)) else x)
```
